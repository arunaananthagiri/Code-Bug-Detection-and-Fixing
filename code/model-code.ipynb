```python
def check_syntax(code):
    """Checks for syntax errors and returns an error message if found."""
    try:
        compile(code, "<string>", "exec")
        return None  # No error found
    except SyntaxError as e:
        return f"Syntax Error on line {e.lineno}: {e.msg}"
    except Exception as e:
        return f"Error: {str(e)}"

def generate_fixed_code(code):
    """Uses CodeGen to suggest a corrected version of the input code."""
    prompt = f"### Fix the following Python code:\n{code}\n### Corrected Code:\n"
    inputs = tokenizer(prompt, return_tensors="pt").to("cuda" if torch.cuda.is_available() else "cpu")

    with torch.no_grad():
        outputs = model.generate(**inputs, max_length=512)

    fixed_code = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # Extract only the corrected code
    corrected_code = fixed_code.split("### Corrected Code:")[-1].strip()
    return corrected_code
```

```python
def main():
    print("Enter your Python code (end input with CTRL+D or an empty line):")

    # Get multi-line code input
    code_lines = []
    while True:
        try:
            line = input()
            if not line.strip():  # Stop on empty line
                break
            code_lines.append(line)
        except EOFError:
            break  # Stop on CTRL+D

    code = "\n".join(code_lines)

    # Check for syntax errors
    error_message = check_syntax(code)

    if error_message:
        print("\n❌ Bug Detected:")
        print(error_message)
        print("\n🔧 Suggested Fix:")
        corrected_code = generate_fixed_code(code)
        print(corrected_code)
    else:
        print("\n✅ No error found!")


if __name__ == "__main__":
    main()
```

```python
import torch
import traceback
from transformers import AutoModelForCausalLM, AutoTokenizer
from safetensors.torch import load_file

model_name = "Salesforce/codegen-2B-multi"
model_path = "/content/drive/MyDrive/intel/code_debugger_model/model-00001-of-00002.safetensors"

# Load tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_name)

# Load model architecture
model = AutoModelForCausalLM.from_pretrained(model_name, state_dict=None, device_map="auto")

# Load weights from safetensors
state_dict = load_file(model_path)
model.load_state_dict(state_dict, strict=False)
```

```python
def execute_code(code):
    """Runs the code and captures errors."""
    try:
        exec(code, {})
        return None  # No error
    except Exception as e:
        return traceback.format_exc()

def generate_fixed_code(code):
    """Uses CodeGen to suggest a corrected version of the input code."""
    prompt = f"### Fix the following Python code:\n{code}\n### Corrected Code:\n"
    inputs = tokenizer(prompt, return_tensors="pt").to("cuda" if torch.cuda.is_available() else "cpu")

    with torch.no_grad():
        outputs = model.generate(**inputs, max_length=512)

    fixed_code = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # Extract only the corrected code
    corrected_code = fixed_code.split("### Corrected Code:")[-1].strip()
    return corrected_code
```

```python
def main():
    print("Enter your Python code (end input with CTRL+D or an empty line):")

    # Get multi-line code input
    code_lines = []
    while True:
        try:
            line = input()
            if not line.strip():  # Stop on empty line
                break
            code_lines.append(line)
        except EOFError:
            break  # Stop on CTRL+D

    code = "\n".join(code_lines)

    # Try executing the code
    error_message = execute_code(code)

    if error_message:
        print("\n❌ Bug Detected:")
        print(error_message)
        print("\n🔧 Suggested Fix:")
        corrected_code = generate_fixed_code(code)
        print(corrected_code)
    else:
        print("\n✅ No error found!")

if __name__ == "__main__":
    main()
```

```python
def check_for_errors(user_code):
    """
    Analyzes the given Python code for errors.
    Returns:
        - "No error found" if the code is correct.
        - Error description if an issue is detected.
    """
    try:
        compile(user_code, "<string>", "exec")  # Check syntax errors
    except SyntaxError as e:
        return f"Syntax Error: {e.msg} at line {e.lineno}, column {e.offset}"
    except Exception as e:
        return f"Runtime Error: {str(e)}"

    return "No error found!"  # No syntax issues

def fix_code(user_code):
    """
    Uses the LLM model to correct the given code.
    """
    prompt = f"### Debug and fix this Python code:\n{user_code}\n### Fixed Code:\n"
    inputs = tokenizer(prompt, return_tensors="pt").to("cuda")  # Send to GPU if available
    outputs = model.generate(**inputs, max_new_tokens=200)
    fixed_code = tokenizer.decode(outputs[0], skip_special_tokens=True)

    return fixed_code.split("### Fixed Code:")[-1].strip()
```

```python
if __name__ == "__main__":
    print("Enter your Python code (end input with CTRL+D or an empty line):")
    user_code = []

    while True:
        try:
            line = input()
            if not line.strip():  # Stop on empty line
                break
            user_code.append(line)
        except EOFError:
            break

    user_code = "\n".join(user_code)

    # Check for errors
    error_message = check_for_errors(user_code)

    if error_message == "No error found!":
        print("✅ No error found!")
    else:
        print(f"❌ Error Detected:\n{error_message}\n")
        print("🔧 Fixing the code...\n")
        corrected_code = fix_code(user_code)
        print(f"✅ Corrected Code:\n{corrected_code}")
```
